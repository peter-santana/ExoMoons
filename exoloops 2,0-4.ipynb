```python
import exomlib
import numpy as np
import matplotlib.pyplot as plt

# make plots inline
%matplotlib inline
```

```python
# star properties
mass_star = 0.99 # solar masses

# planet properties
planet_name = "HD 4113 b"
mass_planet = 1.56 # jupiter masses
a_planet = 1.28 # AU, astronomical units
e_planet = 0.9 # eccentricity
planet_radius = (mass_planet**0.5)

# moon properties
e_moon = 0 # moon eccentricity   
a_moon = 1 #lunar distance
mass_moon = 1 # earth masses

n_orbits = 1000 #Number of orbits
n_points_orbit = 100 #Number of points per orbit
```

```python
# Distance and Mass Variator
while a_moon<=5: 
    while mass_moon<=5:
        data = exomlib.exomoon_orbit1(mass_star = mass_star, mass_planet = mass_planet, 
                                      a_planet = a_planet, e_planet = e_planet, mass_moon = mass_moon, 
                                      a_moon = a_moon, e_moon = e_moon, n_orbits = n_orbits, 
                                      n_points_orbit = n_points_orbit)
        #Figure 1
        fig = plt.figure(figsize = (15,10))
        ax1 = plt.subplot(211)
        # ax1.set_xlabel('Time (years)')
        ax1.set_ylabel('Planet semi-major axis (AU)')
        plt.plot(data.t/(2*np.pi), data.a);
        ax2 = plt.subplot(212, sharex = ax1)
        ax2.set_xlabel('Time (years)')
        ax2.set_ylabel('Moon distance (lunar distances)')
        plt.plot(data.t/(2*np.pi), data.d);
        plt.subplots_adjust(hspace = 0.0)
        plt.savefig('Planet SMA and Moon Distance')
        
        #Figure2
        fig = plt.figure(figsize = (5,5))
        ax = plt.subplot(111)
        ax.set_xlabel('Distance (AU)')
        ax.set_ylabel('Distance (AU)')
        plt.plot(data.xs, data.ys, color = 'black');
        plt.plot(data.x, data.y, color = 'red', label = 'moon');
        plt.plot(data.xp, data.yp, color = 'blue', label = 'planet');
        plt.legend(frameon = False);
        plt.savefig('Orbits')
        
        mass_moon+=1 #step increase
        
    a_moon+=1 #step increase
    mass_moon = 1 #reset steps
```
